- Add quadratic terms for all the numeric variables in the fish example. For example
- X['L1squared'] = np.square(df.Length1)
-  Build a model and find R^2 and RMSE.

```python
import pandas as pd
from pandas.plotting import scatter_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

from sklearn.model_selection import train_test_split

df = pd.read_csv("../data/fish.csv")
print(df.head())

scatter_matrix(df
               )
```

```python
X = df.drop("Weight", axis = 'columns')
x = pd.get_dummies(X)
y = df.Weight

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1)
model = LinearRegression()
model.fit(X_train, y_train)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("r2_score:", model.score(X_train, y_train))

y_hat = model.predict(X_test)
print("rmse: ", root_mean_squared_error(y_test, y_hat))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

X['L1squared'] = np.square(df.Length1)
X['L2squared'] = np.square(df.Length2)
X['L3squared'] = np.square(df.Length3)
X['Hsquared'] = np.square(df.Height)
X['Wsquared'] = np.square(df.Width)

```

```python
model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
model.fit(X_train, y_train)
yhat = model.predict(X_test)


print('R squared:', model.score(X_train,y_train))
print('RMSE', mean_squared_error(y_test, yhat, squared=False))
```
